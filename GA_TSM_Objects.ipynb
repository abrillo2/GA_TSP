```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#gene defination
class Node:
    
    def __init__(self,x,y):
        
        self.x = x
        self.y = y
    
    
    def distanceToNode(self,node):
        
        return self.euclideanDistance(node)
    
    def euclideanDistance(self,node):
        
        distance = np.sqrt((self.x - node.x)**2 + (self.y - node.y)**2)
        return distance
    
    #get x,y property of nodes for crossover
    def getXY(self):
        
        return self.x,self.y
    
    def __repr__(self):
        
        return "{" + str(self.x) + "," + str(self.y) + "}"
    
#populatin fitness(route one divided by distance b/w each node for 1 round trip)
class Fitness:
    
    def __init__(self,route):
        
        self.route = route
        self.distance = 0#integer
        self.fitness = 0.0#float nuber
    
    #route round trip distance calculator
    def roundTripDistance(self):
        
        distance = 0
        
        for i in range(len(self.route)-1):
            
            fromNode = self.route[i]
            toNode = self.route[i+1]
            
            distance+= fromNode.distanceToNode(toNode)
        distance+= self.route[-1].distanceToNode(self.route[0])
        
        self.distance = distance
        
        return self.distance
        
    #route fitness calculator
    
    def routeFitness(self):
        
        self.roundTripDistance()
        
        self.fitness =  1.0/ float(self.distance)
        
        return self.fitness
    
    #returns number of genes
    def geneSize(self):
        
        return len(self.route)
    

        
        
        
```
